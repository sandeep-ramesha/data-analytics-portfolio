```python
# Assignment
on insurance data
1) each region wise total expenses
2) gender wise avg bmi and expenses
3) each region,each children and smoker class wise total expenses

use: group by and pivot table
```

```python
ins = pd.read_csv("insurance.csv")
```

```python
ins
```

```python
ins.groupby(by ="region")['expenses'].sum().sort_values(ascending=False)
```

```python
ins.groupby(by="sex")[['bmi','expenses']].mean().round(2) #using round() to round off decimal 
```

```python
ins.groupby(by = ['region','children','smoker'])['expenses'].sum()
```

```python
ins.groupby(by = ['region','children','smoker'])['expenses'].mean()
```

```python
pd.pivot_table(data=ins,values='expenses',index='region',columns=['smoker','children'],aggfunc='mean').round(2)
```
