```python
What is Data Visualization?
it is the process of representing data in a graphical or visual format such as charts, graphs or maps 
so that people can easily understand patterns, trends and insights.

why visualization is important?
* makes data easy to understand
* helps in quick decision making
* identifies patterns and trends 
* detects outliers or erros 
* improves communication of insights 
```

```python
Univariate Analysis:
    it means analyzing only one variable at a time. to check the quality of the data.
    purpose: 
    * understand distribution
    * find patterns 
    * detect outliers
    ex: histogram, Box plot, KDE plot, Violin plot
    Realtime ex: analyze salary distribution in a company
                -> are most employees low paid or high paid


Bivariate Analysis:
    it means analyzing the relationship between two varibales.
    purpose:
    * understand relationship
    * identify correlation
    * compare variables
    ex: scatter plot, bar plot, line plot
    ex: analyze advertising spend Vs scales
        Does increasing ads increses sales?

Multivariate Analysis:
    it means analyzing relationship between more than two variables.
    purpose:
    * understand complex relationships
    * identify patterns across multiple variables
    * perform deep data analysis (EDA)
    EDA: exploratory data analysis 
    ex: Pair plot, heatmap, facet grid, bubble plot
    ex: analyze sales based on region, time and product category
    -> which combination gives highest profit?
```

```python
# matplotlib and seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df
```

```python
histogram:
* shows distribution of data
* helps to detect skewness and spread
```

```python
#distribution of total bill
plt.hist(df['total_bill'])
plt.title("histogram")
plt.xlabel("Total Bill")
plt.ylabel("frequency")
plt.show()
```

```python
df['total_bill'].max()
```

```python
df['total_bill'].min()
```

```python
df['total_bill'].mean()
```

```python
df['total_bill'].median()
```

```python
#Density Plot
KDE plot:
* smooth version of histogram
* better for understanding data density
```

```python
plt.figure()
sns.kdeplot(df['total_bill'],fill=True)
plt.title("KDE Plot")
plt.xlabel("Total Bill")
plt.ylabel("frequency")
plt.show()
```

```python
box plot:
* identifies outliers
* shows median,quartiles, spread 
```

```python
plt.figure()
sns.boxplot(x=df['total_bill'])
plt.title('Box Plot')
plt.xlabel('Total Bill')
plt.show()
```

```python
Violin plot:
* combines box plot + distribution
* shows data shape clearly 
```

```python
plt.figure()
sns.violinplot(x=df['total_bill'])
plt.title("violin plot")
plt.show()
```

```python
Bar Plot:
* compares values across categories
* easy to understand differences
```

```python
plt.figure()
sns.barplot(x="day", y="total_bill",data=df)
plt.title("bar plot")
plt.show()
```

```python
pie chart:
* shows proportion/percentage
* good for quick insights
note: use only for small number of categories
```

```python
plt.figure()
df['day'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("pie chart")
plt.show()
```
