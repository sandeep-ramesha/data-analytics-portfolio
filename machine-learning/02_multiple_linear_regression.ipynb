```python
x = cars.drop('MPG',axis=1)
```

```python
y = cars['MPG']
```

```python
mlr = LinearRegression()
mlr.fit(x,y)
y_pred = mlr.predict(x)
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,y_pred)
```

```python
error = y - y_pred
```

```python
import seaborn as sns
```

```python
sns.kdeplot(error) #Checking normality
sns.show()
```

```python
Error is not normally distributed, it is right skewed 
```

```python
plt.scatter(cars['HP'],error) #Now checking for HOMOSCEDASTICITY
plt.show()
```

```python
plt.scatter(cars['SP'],error)
plt.show()
```

```python
plt.scatter(cars['WT'],error)
plt.show()
```

```python
plt.scatter(cars['VOL'],error)
plt.show()
```

```python
Error is not uniformly distributed so there is no homoscadasticity
```
