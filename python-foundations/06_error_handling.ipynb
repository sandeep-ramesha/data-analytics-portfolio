```python
error(also called as exception) are issues that stop your program from running properly, 
error handeling means detecting and managing those errors gracefully instead of letting the program crash.
1) Syntax Error
2) Name Error
3) Indent Error
4) type error
5) zerodivision error
6) index error 
7) value error
8) key error
9) filenotfound error 
```

```python
syntax of exception

try:
    # code that might cause an error
except:
    # code to handle the error 
```

```python

x = int(input("Enter a number"))
print(10/x)
```

```python
try:
    x = int(input("Enter a number"))
    print(10/x)
except:
    print("Something went wrong")
```

```python
Catching specific exceptions:
```

```python
try:
    a = int(input("Enter a value of a"))
    b = int(input("Enter a value of b:"))
    c = a/b
    print("The answer of divide by b",c)
except ValueError:
    print("Entered value is wrong")
except ZeroDivisionError:
    print("Can't divide by zero")
except:
    print("Something went wrong")
```

```python
Handle multiple exception with a single except clause:
```

```python
try:
    a = int(input("Enter a value of a"))
    b = int(input("Enter a value of b:"))
    c = a/b
    print("The answer of divide by b",c)
except(ValueError, ZeroDivisionError):
    print("Please enter a valid value")
```

```python
Clean-up actions using finally:

try:
    #bloack of code
    #this may throw an exception
finally:
    #block of code
    #this will always be executed 
    #after the try and any except block
```

```python
try:
    a = int(input("Enter a value of a"))
    b = int(input("Enter a value of b:"))
    c = a/b
    print("The answer of divide by b",c)
except ZerodivisionError:
    print("Can't divide with Zero")
finally:
    print("Inside a finally block")
    
```

```python
Using try, exception with else clause:
```

```python
try:
    # block of code
except Exception1:
    #block of code
else:
    #this code executes when exceptions not occured

try: The try block for risky code that can throw an exception.
except: The except block to handle error raised in a try block.
else: The else block is executed if there is no exception.
```

```python
try:
    a = int(input("Enter a value of a"))
    b = int(input("Enter a value of b:"))
    c = a/b
    print("The answer of divide by b",c)
except ZerodivisionError:
    print("Can't divide with Zero")
else:
    print("We are in else block")
    
```
