```python
#WAP to check if a given number is prime or not
#15 = 2,3,4,....14= 3,5
#17 = 2,3,4,....16= its a prime
#General method

n=int(input("Enter a number:"))
count = 0
for i in range(2,n,1):
    if(n%i==0):
        count=count+1
if(count==0):
    print(n,"is a prime number")
    print(f"it is divided by {count} other numbers")
else:
    print(n,"is not a prime number")
    print(f"it is divided by {count} other numbers")
```

```python
#Check that given number is palindrome or not, eg 1221 is palindrome, 1234 is not palindrome 

n = int(input("Enter a number"))
temp = n
rev = 0

while(n>0):
    rem = n%10
    rev = (rev*10)+rem
    n = n//10
print(f"temp={temp},rev={rev}")
if(temp==rev):
    print("its palindrome")
else:
    print("its not a palindrome")



```

```python
# While inside while

i=1
while(i<=5):
    j = 5
    while(j>=i):
        print(" ",end="")
        j -= 1
    k=1
    while(k<=i):
        print("*",end=" ")
        k += 1
    print()
    i += 1

x = 4
while(x>=1):
    y = 5
    while(y>=x):
        print(" ",end="")
        y -= 1
    z = 1
    while(z<=x):
        print("*",end=" ")
        z+=1
    print()
    x-=1
```
