```python
s1 = 'python123@#5*9.,'
#create seperate list for char,num,sp char

cl = []
nl = []
scl = []
for i in s1:
    if i.isalpha():
        cl.append(i)
    elif i.isdigit():
        nl.append(i)
    else:
        scl.append(i)
print(f"{cl}\n{nl}\n{scl}")
```

```python
s2 = "python is best programming"
#count num of vowels and consonents
v = 0
c = 0
for i in s2:
    if(i=='a' or i=='e' or i=='i' or i=='o' or i=='u'):
        v+=1
    else:
        c+=1
print(f"vowels count is {v}")
print(f"consonents count is {c}")
```

```python
s2 = "python is best programming"
vc = 0
cc = 0
for i in s2:
    if i!=" ":
        if i in "aeiouAEIOU":
            vc+=1
        else:
            cc+=1
print(f"vowel count is {vc}\nconsonents count is {cc}")
```

```python
Create a new string made of the first, middle , and last characters of each input string. Given two strings, s1 and s2,
write a program to return a new string made of s1 and s2 s first, middle and last characters.
s1 = "America"
s2 = "Japan"
Expected output--> AJrpan
```

```python
s1 = "America"
s2 = "Japan"
l1 = len(s1)//2
a = s1[0]
b = s1[l1]
c = s1[-1]

l2 = len(s2)//2
x = s2[0]
y = s2[l2]
z = s2[-1]
op = "".join ([a,x,b,y,c,z])
print(op)
```

```python
'''Calculate the sum and average of the digits present in the string, Given a string s1, WAP to return the sum and
average of the digits that appear in the string, ignoring all other charaters.'''

str1 = "Python29@8496"
s = 0
c = 0
for i in str1:
    if i.isdigit():
        s = s+int(i)
        c = c+1
print("The sum is",s)
print("The average is",s/c)
    
```
