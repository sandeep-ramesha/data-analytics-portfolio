```python
if else ladder/contineous if else

if(condition1):
    statements of con1
else:
    if(condition2):
        statements of cond2
    else:
        if(condition3)
            statements of cond3

        ...
    else:
        default statemets
```

```python
#6th Assignment Question

units = int(input("Enter the units used"))
if (units <= 100):
    ba = units * 0.5
elif (units > 100 and units <= 200):
    diff_units = units - 100
    ba = 100*0.5+diff_units*0.75
elif (units>200):
    diff_units = units -200
    ba = 100*0.5+100*0.75+diff_units*1

print("Total bill amount is:",ba)
```
