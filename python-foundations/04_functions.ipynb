```python
#create a funtion to print prime numbers from 1 to 20

def prime1():
    for i in range(2,21,):
        for j in range(2,i,1):
            if(i%j==0):
                break
        else:
            print(i)

prime1()
        
```

```python
def prime2(st,sp):
    for i in range(st,sp,1):
        for j in range(2,i,1):
            if(i%j==0):
                break
        else:
            print(i)

prime2(1,21)
```

```python
def prime3():
    l1=[]
    for i in range(2,21,1):
        for j in range(2,i,1):
            if(i%j==0):
                break
        else:
            l1.append(i)
    return l1
```

```python
prime3() 
```

```python
#create function to reverse the number, with input with return
def rev(n):
    re = 0
    while(n>0):
        rem = n%10
        re = (re*10)+rem
        n = n//10
    print(re)
```

```python
rev(1234)
```

```python
lambda functions:
its a small, anonymous(nameless) function in python 
* defined using the keyword lambda instead of def
* can take any number of arguments(inputs) but must contain only on expression
* expression is automatically returned(no need to use return)

var = lambda arguments: expression
```

```python
#square of given number 
sq = lambda n: n*n
```

```python
sq(5)
```

```python
#sum of 2 numbers 
su= lambda a,b: a+b
```

```python
su(2,3)
```

```python
Why to use lambda?
* useful for short, simple operations
* saves code(one-liner)
* often used with map(), filter(), reduce() or as inline functions.

When not to use lambda?
* for complex logic(use def instead)
* when readability is more important that shortness 
```

```python
Recursive functions:

its a function that calls itself until a condition is met

important parts:
1) base condition - stop recursion
2) recursive call - function calling itself
```

```python
#sum of first 5 numbers:

def sum_ser(n):
    if(n==0):
        return 0
    return n+sum_ser(n-1)
```

```python
sum_ser(5)
```

```python
#factorial of a given number
def fact(n):
    if(n==1):
        return 1
    else:
        return n*fact(n-1)
```

```python
fact(5)
```
